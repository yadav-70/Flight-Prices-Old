#Importing the important libraries 
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
% matplotlib inline

```python
train_df=pd.read_excel('data_train.xlsx')
```

```python
train_df.head()
```

```python
test_df=pd.read_excel('test_set.xlsx')
test_df.head()
```

```python

final_df = pd.concat([train_df, test_df], ignore_index=True)
final_df.head()
```

```python
final_df.tail()
```

```python
final_df.info()
```

```python
#Feature Engineering 
final_df['Date']=final_df['Date_of_Journey'].str.split('/').str[0]
final_df['Month']=final_df['Date_of_Journey'].str.split('/').str[1]
final_df['Year']=final_df['Date_of_Journey'].str.split('/').str[2]
```

```python
final_df.head()
```

```python
final_df['Date']=final_df['Date'].astype(int)
final_df['Month']=final_df['Month'].astype(int)
final_df['Year']=final_df['Year'].astype(int)
```

```python
final_df.info()
```

```python
final_df.drop('Date_of_Journey',axis=1,inplace=True)
```

```python
final_df.head()
```

```python
final_df['Arrival_Time'].str.split(' ').str[0]
```

```python
#Other Option using lambda
final_df['Arrival_Time']=final_df['Arrival_Time'].apply(lambda x: x.split(' ')[0])
```

```python
final_df.head()
```

```python
final_df.isnull().sum()
```

```python
#From Arrivals we take out hours and minutes 
final_df['Arrival_hour']=final_df['Arrival_Time'].str.split(':').str[0]
final_df['Arrival_min']=final_df['Arrival_Time'].str.split(':').str[1]
```

```python
final_df.head()
```

```python
final_df['Arrival_hour']=final_df['Arrival_hour'].astype(int)
final_df['Arrival_min']=final_df['Arrival_min'].astype(int)
```

```python
final_df.info()
```

```python
final_df.drop('Arrival_Time',axis=1,inplace=True)
```

```python
final_df.head()
```

```python
# Assuming final_df is your DataFrame and it has a 'Dep_Time' column
final_df['Dep_hour'] = final_df['Dep_Time'].str.split(':').str[0]
final_df['Dep_min'] = final_df['Dep_Time'].str.split(':').str[1]
```

```python
final_df['Dep_hour']=final_df['Dep_hour'].astype(int)
final_df['Dep_min']=final_df['Dep_min'].astype(int)
```

```python
final_df.drop('Dep_Time',axis=1,inplace=True)
```

```python
final_df.info()
```

```python
final_df['Total_Stops'].unique()
```

```python
final_df['Total_Stops']=final_df['Total_Stops'].map({'non-stop':0,'1 stop':1,'2 stops':2,'3 stops':3,'4 stops':4,'nan':1})
```

```python
final_df.head()
```

```python
final_df.drop('Route',axis=1,inplace=True)
```

```python
final_df.head()
```

```python
final_df['Additional_Info'].unique()
```

```python
#Converting the duration into the minutes
final_df['duration_hour']=final_df['Duration'].str.split(' ').str[0].str.split('h').str[0]
```

```python
#There is a data which contain data into 5m which is disturbing our data to avoid this we will delete the data from the dataset
final_df[final_df['duration_hour']=='5m']
```

```python
final_df.drop([6474, 2660,13343], axis=0, inplace=True)
```

```python
#Converting the data into integers 
final_df['duration_hour']=final_df['duration_hour'].astype(int)
```

```python
final_df.head()
```

```python
#We are getting the values which is minutes 
final_df['duration_min']=final_df['Duration'].str.split(' ').str[0].str.split('m').str[1]
#we are getting the nan value error hence we place 0.
final_df['duration_min'] = final_df['duration_min'].fillna(0)
```

```python
final_df['duration_min']=final_df['duration_min'].astype(int)
```

```python
final_df.head()
```

```python
final_df.drop('Duration', axis=1, inplace=True)
```

```python
final_df.head()
```

```python
final_df['Airline'].unique()
```

```python
from sklearn.preprocessing import LabelEncoder
labelencoder=LabelEncoder()
```

```python
final_df['Airline']=labelencoder.fit_transform(final_df['Airline'])
final_df['Source']=labelencoder.fit_transform(final_df['Source'])
final_df['Destination']=labelencoder.fit_transform(final_df['Destination'])
final_df['Additional_Info']=labelencoder.fit_transform(final_df['Additional_Info'])
```

```python
final_df.shape
```

```python
final_df.head()
```

```python
final_df[['Airline']]
```

```python
import numpy as np
import pandas as pd
from sklearn.preprocessing import OneHotEncoder

# Assuming final_df is your DataFrame and 'Airline' is the column you want to encode
# Replace 'final_df' with the actual DataFrame variable name if it's different
ohe = OneHotEncoder()

# Fit and transform the 'Airline' column
encoded_airline = ohe.fit_transform(np.array(final_df['Airline']).reshape(-1, 1)).toarray()

# Convert the encoded data to a DataFrame
encoded_airline_df = pd.DataFrame(encoded_airline, columns=ohe.get_feature_names_out(['Airline']))

# Reset the index of both DataFrames
final_df.reset_index(drop=True, inplace=True)
encoded_airline_df.reset_index(drop=True, inplace=True)

# Concatenate the DataFrames
final_df = pd.concat([final_df, encoded_airline_df], axis=1)
```
